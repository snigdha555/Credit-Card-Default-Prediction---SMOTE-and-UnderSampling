```python
import pandas as pd

df = pd.read_excel('default_credit.xls')
df.head(5)
```

```python
#importing classification libraries

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.preprocessing import RobustScaler

from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report

from sklearn.metrics import accuracy_score, make_scorer
```

```python
df = df.rename(columns={'default payment next month': 'DEFAULT','PAY_0': 'PAY_1'})

df.head()
```

```python
df1 = df.drop('ID', axis = 1)

df1.head()
```

```python
X = df1.drop('DEFAULT', axis = 1)

y = df1['DEFAULT']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.20,
    random_state=4)

scaler = RobustScaler()

normalized_x_train = pd.DataFrame(
    scaler.fit_transform(x_train),
    columns = x_train.columns
)

LR = LogisticRegression(
    C=0.01,
    solver='liblinear'
).fit(normalized_x_train, y_train)

normalized_x_test = pd.DataFrame(
    scaler.transform(x_test),
    columns = x_test.columns
)
y_test_pred = LR.predict(normalized_x_test)
print('Accuracy:', accuracy_score(y_test,y_test_pred))
```

```python
df_train = pd.concat([x_train,y_train],axis = 1)

df_train.head()
```

```python
ind = df_train[df_train.DEFAULT==1].index

print(len(ind))
```

```python
ind1 = df_train[df_train.DEFAULT==0].index

print(len(ind1))
```

```python
df_minority = df_train.loc[ind]

df_minority.shape
```

```python
df_majority = df_train.loc[ind1]

print(df_majority.shape)

df_majority.head()
```

```python
df_maj_undersampled = df_majority.sample(n=len(ind),random_state = 123)
```

```python
print(df_maj_undersampled.shape)

df_maj_undersampled.head()
```

```python
#merging minority and downsampled dataframes

df_undersampled = pd.concat([df_maj_undersampled,df_minority],axis =0)
```

```python
#shuffling the dataset

from sklearn.utils import shuffle

df_undersampled = shuffle(df_undersampled)

df_undersampled.head()

df_undersampled.shape
```

```python
#new X_train and y_train for downsampled

X_trainNew = df_undersampled.drop('DEFAULT', axis = 1)

y_trainNew = df_undersampled['DEFAULT']
```

```python


scaler = RobustScaler()

normalized_x_train_us = pd.DataFrame(
    scaler.fit_transform(X_trainNew),
    columns = X_trainNew.columns
)

LR_us = LogisticRegression(
    C=0.01,
    solver='liblinear'
).fit(normalized_x_train_us, y_trainNew)


y_test_pred_us = LR_us.predict(normalized_x_test)
print('Accuracy:', accuracy_score(y_test,y_test_pred_us))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF_us = RandomForestClassifier().fit(normalized_x_train_us, y_trainNew)


y_test_pred_us_RF = RF_us.predict(normalized_x_test)
print('Accuracy:', accuracy_score(y_test,y_test_pred_us_RF))
```

```python
print(classification_report(y_test,y_test_pred_us))
```

```python
print(classification_report(y_test,y_test_pred_us_RF))
```
